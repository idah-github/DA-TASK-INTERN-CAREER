```python
# df.select_dtypes(include=[ 'number','float']).head()
categories_cnt=df['Categories'].value_counts()
# categories_cnt
plt.figure(figsize=(10, 7))
sbn.barplot(x=categories_cnt.index, y= categories_cnt.values)
plt.xticks(rotation=90)
plt.title('Popular Categories')
plt.show()
```

```python
#distribution of streamers' audiences by country.
plt.figure(figsize=(13, 7))
sbn.barplot(x='Country', y='Subscribers', data=df)
plt.xticks(rotation=90)
plt.title('Audience Distribution by Country')
plt.show()
#using a piechart
country_subscribers = df.groupby('Country')['Subscribers'].sum()
plt.figure(figsize=(20, 10))
plt.pie(country_subscribers, autopct='%1.1f%%')
plt.title('Audience Distribution by Country')
plt.legend(country_subscribers.index, loc='upper center', bbox_to_anchor=(1.1, 0.8), shadow=True)
plt.show()

```

```python
#Explore the distribution of content categories. Which categories have the highest number of
# subscribers
# plt.figure(figsize=(10, 7))
# sbn.barplot(x='Categories', y='Subscribers', data=df)
# plt.xticks(rotation=90)
# plt.title('Audience Distribution by Content Category')
# plt.show()
# #to the above plot makee the subcribers scale into 1000 for better visualization
df['Subscribers (in thousands)'] = df['Subscribers'] / 1000
plt.figure(figsize=(14, 7))
sbn.barplot(x='Categories', y='Subscribers (in thousands)', data=df)
plt.xticks(rotation=90)
plt.title('Audience Distribution by Content Category')
plt.show()
```
